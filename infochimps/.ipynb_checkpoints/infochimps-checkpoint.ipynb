#  Acquire Parse Filter

 Acquire tags from infochimps.com using link to the tag_spider.py 
 
 scrapy runspider page_spider.py -o tag_pages_export.json -t json 
 
 for example
 {"tags": ["expenditures (59)", "financial (46)", ...] }


 Parse
 tag_pages_export.json is parsed to fetch the frequecy count
 
 for example
 [{"text": "population", "size": 1777}, {"text": "census", "size": 1681}, ...}]

Filter
Summary of scraping tags from infochamps.com using scrapy
1. We found a total of 21193 tags
2. 16259 are tags with no data set/ pages associated with them. 
3. 4934 have atleast 1 page.
4. 169 have more than 50 pages associated with them
5. We have included these 169 tag in this study.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tag_cloud = pd.read_csv('tag_cloud.csv')
tag_cloud.head()
ylabels = tag_cloud['tag']
plot = tag_cloud.plot(kind='barh', figsize=(32, 64))
plot.tick_params(axis='both', which='major', labelsize=15)
plot.set_yticklabels(ylabels)
print 'hello'
```

We see a clear partition in the distribution 
1. greater than 400
2. between 400 and 200
3. rest > 50
