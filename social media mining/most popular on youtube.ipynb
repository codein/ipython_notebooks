```python
from youtube import GetFeedByUrl
from IPython.core.display import Image, display

uri = 'http://gdata.youtube.com/feeds/api/standardfeeds/most_popular'
video_infos = GetFeedByUrl(uri)

for video_info in video_infos:
    display(Image(url=video_info['thumbnail'], embed=True))
    for key, value in video_info.iteritems():
        print '%s: %s' % (key, value)
    
```
